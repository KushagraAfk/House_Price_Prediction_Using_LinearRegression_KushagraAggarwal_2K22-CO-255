```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
train = pd.read_csv('train.csv')
```

```python
test = pd.read_csv('test.csv')
```

```python
train.drop(columns = ['ID'], inplace = True)
```

```python
test.drop(columns = ['ID'], inplace = True)
```

```python
reg = LinearRegression()
```

```python
reg.fit(train.drop(columns = ['medv']), train['medv'])
```

```python
predtrain = reg.predict(train.drop(columns = ['medv']))
```

```python
print('root mean squared error: ',mean_squared_error(train['medv'],predtrain)**0.5)
```

```python
print('r2 score: ',r2_score(train['medv'],predtrain))
```

```python
predtraindf = pd.DataFrame(predtrain, columns = ['medv'])
```

```python
train['predicted'] = predtraindf['medv']
```
